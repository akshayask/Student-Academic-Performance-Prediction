# concatinating both mat and por dataset into result dataset

```python
a = df
b = df1
frames = [a,b]
result = pd.concat(frames,axis=0)
```

```python
result['subject'].value_counts()
```

```python
result.shape
```

```python
result.info()
```

locating rows of G3 that lies between a specified range


```python
result.loc[(result.G3 >= 18) & (result.G3 <= 20), 'FinalGrade'] = 'Excellent'
result.loc[(result.G3 >= 15) & (result.G3 <= 17), 'FinalGrade'] = 'Good' 
result.loc[(result.G3 >= 11) & (result.G3 <= 14), 'FinalGrade'] = 'Satisfactory' 
result.loc[(result.G3 >= 6) & (result.G3 <= 10), 'FinalGrade'] = 'Poor' 
result.loc[(result.G3 >= 0) & (result.G3 <= 5), 'FinalGrade'] = 'Failure' 
result.head(5)
```

```python
result.to_csv("result.csv",index="False")
```

```python
result
```

```python
result.loc[(result.absences >= 0) & (result.absences <= 9), 'Regularity'] = 'Always regular'
result.loc[(result.absences >= 10) & (result.absences <= 29), 'Regularity'] = 'Mostly regular' 
result.loc[(result.absences >= 30) & (result.absences <= 49), 'Regularity'] = 'Regular' 
result.loc[(result.absences >= 50) & (result.absences <= 79), 'Regularity'] = 'Irregular' 
result.loc[(result.absences >= 80) & (result.absences <= 93), 'Regularity'] = 'Highly irregular' 
result.head(20)
result.tail()
```

```python
result.loc[(result.G1 >= 18) & (result.G1 <= 20), 'GradeG1'] = 'Excellent'
result.loc[(result.G1 >= 15) & (result.G1 <= 17), 'GradeG1'] = 'Good' 
result.loc[(result.G1 >= 11) & (result.G1 <= 14), 'GradeG1'] = 'Satisfactory' 
result.loc[(result.G1 >= 6) & (result.G1 <= 10), 'GradeG1'] = 'Poor' 
result.loc[(result.G1 >= 0) & (result.G1 <= 5), 'GradeG1'] = 'Failure' 
result.head(5)
```

```python
result.loc[(result.G2>= 18) & (result.G2 <= 20), 'GradeG2'] = 'Excellent'
result.loc[(result.G2 >= 15) & (result.G2 <= 17), 'GradeG2'] = 'Good' 
result.loc[(result.G2 >= 11) & (result.G2 <= 14), 'GradeG2'] = 'Satisfactory' 
result.loc[(result.G2 >= 6) & (result.G2 <= 10), 'GradeG2'] = 'Poor' 
result.loc[(result.G2 >= 0) & (result.G2 <= 5), 'GradeG2'] = 'Failure' 
result.head(5)
```

```python
perc = (lambda col: col/col.sum())
gradeby_freetime = pd.crosstab(index=result.FinalGrade, columns=result.freetime)
print(gradeby_freetime)
freetime = gradeby_freetime.apply(perc)
print(freetime)
#print(type(freetime))
freetime.plot.bar(fontsize=16, figsize=(14,8))
plt.title('Grade By freetime', fontsize=20)
plt.ylabel('Percentage of Students', fontsize=12)
plt.xlabel('Final Grade', fontsize=16)
plt.show()
```

```python
perc = (lambda col: col/col.sum())
gradeby_studytime = pd.crosstab(index=result.FinalGrade, columns=result.studytime)
print(gradeby_studytime)
studytime = gradeby_freetime.apply(perc)
print(studytime)
#print(type(studytime))
freetime.plot.bar(fontsize=16, figsize=(14,8))
plt.title('Grade By studytime', fontsize=20)
plt.ylabel('Percentage of Students', fontsize=12)
plt.xlabel('Final Grade', fontsize=16)
plt.show()
```

```python
perc = (lambda col: col/col.sum())
gradeby_school = pd.crosstab(index=result.FinalGrade, columns=result.school)
print(gradeby_school)

print("*************************")
school = gradeby_school.apply(perc)
print(school)
#print(type(studytime))
school.plot.bar(fontsize=16, figsize=(14,8))
plt.title('Grade By school', fontsize=20)
plt.ylabel('Percentage of Students', fontsize=12)
plt.xlabel('Final Grade', fontsize=16)
plt.show()
```
